```python
def procesar_json(ruta):
    # Crea una lista para almacenar los datos extraídos
    datos = []
    # Recorre cada archivo en la ruta proporcionada
    for archivo in os.listdir(ruta):
        
        # Si el archivo es un archivo JSON
        if archivo.endswith('.json'):
            print("Procesando archivo:", archivo, "en la carpeta:", ruta)
            # Abre el archivo JSON y extrae los datos
            with open(os.path.join(ruta, archivo)) as f:
                datos_json = json.load(f)
                # Agrega los datos a la lista de datos
                datos.append(datos_json)
                
    # Devuelve la lista de datos
    return datos
```

```python

# Crea una lista para almacenar todos los datos de los archivos JSON
todos_los_datos = []

# Recorre cada subcarpeta en el cuarto nivel y procesa los archivos JSON
for ruta, carpetas, archivos in os.walk(ruta_principal):
    if ruta_principal in carpetas:
        carpetas.remove(ruta_principal)
    if len(os.path.relpath(ruta, ruta_principal).split(os.sep)) == 3: #dentro de 'users' hay 3 niveles de carpetas hasta llegar a los json
        datos = procesar_json(ruta)
        todos_los_datos.extend(datos)
```

```python
# Crea un archivo CSV y escribe los datos en él
with open('datos' + curr_date + '.csv', 'w', newline='') as f:
    escritor_csv = csv.writer(f)
    # Escribe el encabezado del archivo CSV
    escritor_csv.writerow(['userId','sessionId', 'timestamp', 'navigation'])
    # Escribe los datos de cada fila en el archivo CSV
    for datos in todos_los_datos:
        escritor_csv.writerow([datos['userId'].get('value'), datos['sessionId'].get('value'), datos['timestamp'].get('value'), datos['navigation'].get('value')])
```

```python
path_import = '..\datos\marta\datos.csv'
df = pd.read_csv(path_import)
df
```

```python
df = df.sort_values(['sessionId', 'timestamp'], ascending=[True, True])
df.head(20)
```

```python
df = df.rename(columns={"visitId": "a", "navigation": "pagePath"})
df.head()
```

```python
# Separamos la columna pagePath en diferentes columnas
split_path = df['pagePath'].str.split('/', expand=True)[1:4].apply(lambda x: x.str.strip('/'))

# Creamos una lista con el nombre de las nuevas columnas
new_columns = ['pagePathLevel1']

# Añadimos una nueva columna si la ruta tiene más de 3 niveles
if len(df['pagePath'].str.split('/')) > 1:
    split_path[1] = df['pagePath'].str.split('/')[1].apply(lambda x: '/' + x if x != '' else '')

# Asignamos los nombres de las nuevas columnas al dataframe
split_path.columns = new_columns[:split_path.shape[1]]

# Concatenamos el dataframe original con el dataframe con las nuevas columnas
df = pd.concat([df, split_path], axis=1)
df
```

```python
df
```

```python


# Obtenemos el número máximo de niveles en todas las rutas
max_levels = split_path.notnull().sum().max()

# Creamos las nuevas columnas para cada nivel
new_columns = [f'pagePathLevel{i}' for i in range(1, max_levels + 1)]

# Añadimos las nuevas columnas al dataframe
df[new_columns] = split_path

# Agregamos el "/" al final de cada nivel excepto para el último
for col in new_columns[:-1]:
    df[col] = '/' + df[col] + '/'

# Eliminamos el "/" al final del último nivel
df[new_columns[-1]] = '/' + df[new_columns[-1]]

df
```

```python
# Asignamos las columnas al dataframe
df[['pagePathLevel1', 'pagePathLevel2', 'pagePathLevel3', 'pagePathLevel4']] = split_path.iloc[:, 1:5].apply(lambda x: x.str.strip('/'))
df
```

```python


# Creamos columnas para cada nivel del path
df['pagePathLevel1'] = '/' + split_path[0] + '/'
df['pagePathLevel2'] = '/' + split_path[1] + '/'
df['pagePathLevel3'] = '/' + split_path[2] + '/'
df['pagePathLevel4'] = split_path[3]

# Verificamos si el último nivel es vacío y eliminamos el "/"
df['pagePathLevel1'] = df['pagePathLevel1'].apply(lambda x: x[:-1] if x.endswith('/') else x)
df['pagePathLevel2'] = df['pagePathLevel2'].apply(lambda x: x[:-1] if x.endswith('/') else x)
df['pagePathLevel3'] = df['pagePathLevel3'].apply(lambda x: x[:-1] if x.endswith('/') else x)
```
